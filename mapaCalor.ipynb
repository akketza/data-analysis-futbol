```python
import pandas as pd
from mplsoccer import VerticalPitch, Pitch
from statsbombpy import sb
import matplotlib.pyplot as plt
```

```python
#aca se extrae la competiciones
sb.competitions()
#aca se extrae el match

sb.matches(competition_id=43, season_id=106)
```

```python
#trae los eventos del partido
partido = sb.events(match_id=3869151)
#te da la tabla del partido
partido.head()
```

```python
#traer los nombres de los jugadores
partido.player.value_counts()
```

```python
#filtro por jugador
df_jugador = partido[partido['player'] == 'Lautaro Javier Martínez']
#llama al data frame 
df_jugador.shape
```

```python
#las columnas te traen todas los endpoints

#df_messi.columns

#locacion me da un array de x y
#se define x, y, location para poder visualizarlo en columnas

df_jugador[['x', 'y']] = df_jugador.location.apply(pd.Series)
df_jugador[['end_x', 'end_y']] = df_jugador.pass_end_location.apply(pd.Series)
df_jugador[['x', 'y', 'location']]
```

```python
#se genera el mapa difiniedno ax
fig, ax = plt.subplots(figsize=(16,9))
#como se importo arriba Pitch se trae el tipo statsbomb
pitch = VerticalPitch(pitch_type='statsbomb', line_color='white', pitch_color='black',corner_arcs=True)
#se dibuja la cancha
pitch.draw(ax=ax)
#se llama al kde y se especifica los ax x e y, tambien se le puede dar mas parametros
pitch.kdeplot(
    #df_jugador[df_jugador['type'] == 'Pass']['end_x'],
    #df_jugador[df_jugador['type'] == 'Pass']['end_y'],
    df_jugador['x'],
    df_jugador['y'],
    ax=ax,
    levels=100,
    fill=True,
    zorder=-1,
    shade_lowest=True,
    thresh=0,
    cut=4,
    cmap='hot'
)
```
